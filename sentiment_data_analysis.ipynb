# menentukan pertanyaan
siapakah 20 besar user yang paling aktif?
berikan daftar 10 kata paling sering muncul
bagaimanakan rasio sentiment positif, negatif, dan netral?

```python
# import semua library yang digunakan
import numpy as np 
import pandas as pd 
```

```python
data_df = pd.read_csv('./dataset/dataset_predicted_sentiment.csv')
data_df.head(10)
```

```python
user_sentiment_counts = data_df.groupby(['username', 'predict_sentiment']).size().unstack(fill_value=0)
top_users = user_sentiment_counts.sum(axis=1).nlargest(10)
top_users_data = user_sentiment_counts.loc[top_users.index]
```
